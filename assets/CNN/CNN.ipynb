```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
s = np.linspace(-2*np.pi,2*np.pi, 50)
K = np.sinc(s)
plt.plot(s,K)
plt.grid(True)
plt.savefig('kernel.png')
```

```python
x = np.zeros(50*6)
x[10:60] = K
x[230:280] = K
x += np.random.randn(50*6)*0.01
x[150] = 1.1
x[200] = -0.5
```

```python
plt.plot(x)
plt.grid(True)
plt.ylabel('x')
plt.xlabel('t')
plt.savefig('signal.png')
```
