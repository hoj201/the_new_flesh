```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
np.random.seed(42)
x = np.array([2*np.random.uniform()-1.0 for _ in range(50)])
def decision_func(x):
    if x > 0.0:
        return 1.0
    return 0.0

y = np.array([decision_func(_) for _ in x])
for _ in range(2):
    i = np.random.choice(50)
    y[i] = 1.0 - y[i]
```

```python
plt.scatter(x,y)
plt.grid(True)
plt.savefig('dataset.png')
```

```python
y_hat = np.array(list(map(decision_func,x)))
plt.scatter(x,y,label='raw data')
plt.plot(x,y_hat,'r+',label='logistic')
plt.grid(True)
plt.savefig('dataset_and_predictor.png')
plt.legend()
```

```python
y_hat = np.array(list(map(decision_func,x)))
plt.scatter(x,y,label='raw data')
plt.plot(x,y,'r+',label='naive prediction')
plt.grid(True)
plt.savefig('dataset_and_naive.png')
plt.legend()
```

```python
def decision_func_2(x):
    if x > 0.5 or x < -0.5:
        return 1.0
    return 0.0

np.random.seed(42)
x = np.array([2*np.random.uniform()-1.0 for _ in range(50)])

y = np.array([decision_func_2(_) for _ in x])
for _ in range(2):
    i = np.random.choice(50)
    y[i] = 1.0 - y[i]
    
plt.scatter(x,y)
plt.grid(True)
plt.savefig('dataset2.png')
```

```python
def logistic_regr(x):
    if x < -0.5:
        return 1.0
    return 0.0
y_hat = np.array(list(map(logistic_regr,x)))
plt.scatter(x,y)
plt.plot(x,y_hat,'r+')

plt.grid(True)
plt.savefig('logistic_2.png')
```
